# **16. Classify Movies Based on Ratings [Excellent, Good, and Average]**

```python
data.columns
```

```python
def rating(Rating):
  if Rating>=7.0:
    return "Excellent"
  elif Rating>=6.0:
    return "Good"
  else:
    return "Average"
```

```python
data['Rating_Category']=data['Rating'].apply(rating)
```
