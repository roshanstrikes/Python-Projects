```python
## how to clean the 'name' data containing extras:    (We can create a function)
```

```python
import re
def clean_rest_name(names):
    if isinstance(names,str):
        if re.findall(r"[a-zA-Z0-9 ]", names):
            str1= re.findall(r"[a-zA-Z0-9 ]", names)
            str2= "".join(str1)   ## converting list to string
            return str2
        
    return names
```

```python
 # df['name'].apply(clean_rest_name)
    
df['name']= df['name'].apply(clean_rest_name)
```

```python
def clean_rates(rates):
    if (rates== "NEW" or rates== "-"):
        return np.nan
    else:
        rates_sp= float(str(rates).split("/")[0])
        return rates_sp
```

```python
df['rate']= df['rate'].apply(clean_rates)
```
